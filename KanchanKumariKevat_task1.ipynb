```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/retail_sales_dataset.csv')
```

```python
print("---  (Columns aur Rows) ---")
print(df.info())

print("\n---  (Missing Values)  ---")
print(df.isnull().sum())


df['Date'] = pd.to_datetime(df['Date'])
```

```python
print("Mean,Median,Min,Max")
df.describe()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Customers Age Distribution')

plt.subplot(1, 3, 2)
sns.barplot(x='Product Category', y='Total Amount', data=df, estimator=sum, errorbar=None, palette='pastel')
plt.title('Sales by Product Category')
plt.xticks(rotation=15)

plt.subplot(1, 3, 3)
df_trend = df.groupby('Date')['Total Amount'].sum().reset_index()
plt.plot(df_trend['Date'], df_trend['Total Amount'], color='green', marker='.')
plt.title('Sales Trend Over Time')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(6, 4))
sns.boxplot(x='Gender', y='Total Amount', data=df, palette='Set2')
plt.title('Spending Behavior by Gender')
plt.show()

```
